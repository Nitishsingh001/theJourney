# Pandas Series

### creating Series

#### There are multiple ways to create a pandas Series:

```python
import numpy as np 
import pandas as pd 
```

```python
labels = ['a','b','c']
my_lists = [10,20,30]
arr = np.array([10,20,30])
d = {1:10,2:20,3:30}
```

```python
pd.Series(my_lists)
```

```python
pd.Series(my_lists,index = labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(arr)
```

```python
pd.Series(d)
```

### Creation a Dataframe

```python
data =  {
    'Name' : ['john', 'Anna', 'peter', 'Linda'],
    'Age' : [28,34,29,42], 
    'City' : ['New York', 'Paris', 'Berlin', 'London'], 
    'Salary' : [65000,70000,62000,85000]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
data_list = [
    ['john',28,'New York', 650000], 
    ['Anna', 34, 'paris', 70000], 
    ['peter', 29, 'Berlin', 62000], 
    ['Linda', 42, 'London', 85000]
]
df2 = pd.DataFrame(data_list)
```

```python
df2
```

```python
columns = ["Name",'Age',"City","Salary"]
df2 = pd.DataFrame(data_list, columns = columns)
df2
```
