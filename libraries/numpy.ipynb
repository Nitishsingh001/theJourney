### Numpy Indexing and Slicing of Matrix

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[0]
```

```python
slice = arr[2:6,0:5]
slice
```

```python
slice =  arr[0:2, 3:]
slice
```

```python
col = arr[:,4]
col
```

### Broadcastind 

```python
l1 = [2,4,5,6,7]
arr = np.array(l1)
```

```python
arr + 10
```
